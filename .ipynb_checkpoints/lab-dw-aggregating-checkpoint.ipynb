In this challenge, we will continue to work with customer data from an insurance company. We will use the dataset called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
# Importing the data

import pandas as pd
import numpy as np

#pd.options.display.max_columns = None
#pd.options.display.max_rows = None

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")
display(df)
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
new_df=df[(df["Total Claim Amount"]>1000)&(df["Response"]=="Yes")]
new_df
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# By Gender

df[df["Response"]=="Yes"].groupby(["Gender"]).agg({"Total Claim Amount": "mean"})
```

```python
# By Policy type

df[df["Response"]=="Yes"].groupby(["Policy Type"]).agg({"Total Claim Amount": "mean"})
```

```python
# By Gender/Policy Type

df[df["Response"]=="Yes"].groupby(["Policy Type", "Gender"]).agg({"Total Claim Amount": "mean"})
```

```python
# Creating bar chart for analysis

import matplotlib.pyplot as plt

average_claim_policy_gender = df[df["Response"]=="Yes"].groupby(["Gender","Policy Type"]).agg({"Total Claim Amount": "mean"})

# Reset column names for clarity
average_claim_policy_gender.columns = ["Average Total Claim Amount"]

# Plot the bar chart with different colors and data labels
ax = average_claim_policy_gender.plot(kind='bar', figsize=(10, 6))
ax.set_xlabel('Gender×Policy')
ax.set_ylabel('Average Total Claim Amount')
ax.set_title('Average Total Claim Amount by Gender/Policy')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Set the y-label above 300
ax.set_ylim(300, 500)

plt.show()
```

**Conclusion**
<br>
- For both men and women, the average total claims amount for clients with "corporate auto insurance" is relatively low compared to other clients, especially lower for men. 
- This means that the latest campaign may not have been attractive to this segment. On the other hand, depending on the context, this segment could be considered as a relatively low-risk group.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
df_state = df.groupby(["State"]).agg({"Customer": "count"})
df_state[df_state["Customer"]>500]
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# By Gender

df.groupby(["Gender"]).agg({"Customer Lifetime Value": ["max", "min", "median"]})
```

```python
# By Education

df.groupby(["Education"]).agg({"Customer Lifetime Value": ["max", "min", "median"]})
```

```python
# By Education and Gender

df_clv = df.groupby(["Education", "Gender"]).agg({"Customer Lifetime Value": ["max", "min", "median"]})
df_clv
```

**Conclusion**

- There is a certain relationship between education and gender and customer lifetime value (CLV).
- Customers categorized as "high school and below" appear to have relatively higher CLV than other groups. Note that the variation in values is greater for males than for females, suggesting that some customers may be making a particular contribution.
- The highly educated groups of Doctor and Master have a relatively low CLV, but the variation is small, so marketing products targeted to this group could be an option to consider.
